```python
# Import needed libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read CSV file to dataframe
df = pd.read_csv("Suomalaiset neulojat.csv")
```

```python
# Missing values
nan_count = df.isna().sum()
print(nan_count)
```

```python
# Standardize some of the answers
df['Millaista "neulesomea" kulutat netissä?'] = df['Millaista "neulesomea" kulutat netissä?'].str.replace('Youtube-videot', 'Youtube')
df['Millaista "neulesomea" kulutat netissä?'] = df['Millaista "neulesomea" kulutat netissä?'].str.replace('Instagram-tilit', 'Instagram')
df['Sukupuoli'] = df['Sukupuoli'].apply(lambda x: x if x in ['Nainen', 'Mies'] else 'Muu')
```

```python
df.to_csv('Suomalaiset neulojat.csv', index=True)
```
